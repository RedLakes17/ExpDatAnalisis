```python
import seaborn as sns
import pandas as pd

# Cargar el dataset
tips = sns.load_dataset("tips")

# Ver las primeras filas
print(tips.head())
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex')
```

```python
#Tablas de datos
print(tips['day'].value_counts(normalize=True))
pd.crosstab(tips['day'], tips['time'])
```

```python
pd.crosstab(tips['day'], tips['time'], values=tips['tip'], aggfunc='median')
```
